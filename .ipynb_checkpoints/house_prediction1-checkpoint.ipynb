```python
import seaborn as sns
import matplotlib.pyplot as plt

def compare_countplots(old_data, new_data, feature):
    
    # Create subplots
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Plot count plot for the old feature
    sns.countplot(x=feature, data=old_data, ax=axes[0])
    axes[0].set_title(f'Old Data - Count Plot - {feature}')

    # Plot count plot for the new feature
    sns.countplot(x=feature, data=new_data, ax=axes[1])
    axes[1].set_title(f'New Data - Count Plot - {feature}')

    # Adjust layout
    plt.tight_layout()

    # Show the plots
    plt.show()

# Example usage:
# compare_countplots(old_data, new_data, 'your_feature')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def compare_lotfrontage(old_data, new_data, figsize=(16, 5)):
    """
    Generate side-by-side box plots and histograms for old and new 'LotFrontage' values.

    Parameters:
    - old_data: DataFrame containing old 'LotFrontage' values.
    - new_data: DataFrame containing new 'LotFrontage' values.
    - figsize: Tuple, optional, default: (16, 5). Size of the figure.

    Returns:
    - None (displays the plots).
    """
    plt.figure(figsize=figsize)

    # Box Plot
    plt.subplot(1, 2, 1)
    sns.boxplot(x="LotFrontage", data=old_data)
    plt.title('Old Data - Box Plot - LotFrontage')

    plt.subplot(1, 2, 2)
    sns.boxplot(x="LotFrontage", data=new_data)
    plt.title('New Data - Box Plot - LotFrontage')

    # Adjust layout
    plt.tight_layout()

    # Show the plots
    plt.show()

    # Histogram
    plt.figure(figsize=figsize)

    plt.subplot(1, 2, 1)
    sns.histplot(data=old_data, x="LotFrontage", kde=True)
    plt.title('Old Data - Histogram - LotFrontage')

    plt.subplot(1, 2, 2)
    sns.histplot(data=new_data, x="LotFrontage", kde=True)
    plt.title('New Data - Histogram - LotFrontage')

    # Adjust layout
    plt.tight_layout()

    # Show the plots
    plt.show()

# Example usage:
# compare_lotfrontage(df, df_new)
```

### Model Training

```python

from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.isotonic import IsotonicRegression 

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from xgboost import XGBRegressor

```

```python
svr = SVR()
lr = LinearRegression()
sgd = SGDRegressor()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor()
iso = IsotonicRegression()
gb = GradientBoostingRegressor()
rf = RandomForestRegressor()
mlp = MLPRegressor()
xgb = XGBRegressor()
```

```python
models_dict = {
    'a': ('SVR', svr),
    'b': ('Linear Regression', lr),
    'c': ('SGD Regressor', sgd),
    'd': ('KNeighbors Regressor', knn),
    'e': ('Decision Tree Regressor', dt),
    'f': ('Isotonic Regression', iso),
    'g': ('Gradient Boosting Regressor', gb),
    'h': ('Random Forest Regressor', rf),
    'i': ('MLP Regressor', mlp),
    'j': ('XGBoost Regressor', xgb)
}
```

```python
from sklearn.model_selection import KFold,  cross_val_score
from sklearn.metrics import make_scorer, r2_score

def test_model(model, X_train=X_train, y_train = y_train):
    cv = KFold(n_splits = 7, shuffle = True, random_state = 45)
    r2 = make_scorer(r2_score)
    r2_val_score = cross_val_score(model, X_train, y_train, cv = cv, scoring = r2)
    score = [r2_val_score.mean()]
    return score
```

```python
models_score = []

for model in models_dict:
    if models_dict[model][0] == 'Isotonic Regression':
        continue  # Skip Isotonic Regression
    print("Training model:", models_dict[model][0])
    score = test_model(models_dict[model][1], X_train, y_train)
    print("Score of model:", score)
    models_score.append([models_dict[model][0], score[0]])
    
```
